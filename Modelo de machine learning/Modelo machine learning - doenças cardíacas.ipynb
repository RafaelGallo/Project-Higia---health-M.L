# Treio e teste

```python
x = df.iloc[:, 0:13].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 25)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

# PreparaÃ§Ã£o dos dados

```python
from sklearn.preprocessing import StandardScaler

Standard_scaler = StandardScaler
Standard_scaler = Standard_scaler().fit(x_train)
scaler_train = Standard_scaler.transform(x_train)
scaler_train
```

```python
from sklearn.preprocessing import MinMaxScaler

MinMax = MinMaxScaler().fit(x_train)
MinMax = MinMax.transform(x_train)
MinMax
```

# Modelo de machine learning - Naive bayes

```python
from sklearn.naive_bayes import GaussianNB

naive = GaussianNB()
naive.fit(x_train, y_train)
naive_score = naive.score(x_train, y_train)
naive_score
```

```python
#PrevissÃ£o do naive bayes

naive_pred = naive.predict(x_test)
naive_pred
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(y_test, naive_pred)
matrix_naive = metrics.confusion_matrix(y_test, naive_pred)
print("AcurÃ¡cia Naive bayes\n %.2f" % (acuracia))
print("Matrix de confusÃ£o")
print(matrix_naive)
```

# Modelo de machine learnig - RegressÃ£o logistica

```python
from sklearn.linear_model import LogisticRegression

reg_lor = LogisticRegression()
reg_lor.fit(x_train,y_train)
score = reg_lor.score(x_train, y_train)
reg_pred = reg_lor.predict(x_test)
reg_pred
```

```python
print("Score regressÃ£o logistica: %.2f" % (score * 100))
```

```python
#AcurÃ¡cia

acuracia = accuracy_score(y_test, reg_pred)
matrix  =metrics.confusion_matrix(y_test, reg_pred)

print("AcurÃ¡cia da regressÃ£o logistica\n %.2f" % (acuracia))
print("Matrix da regressÃ£o logistica\n")
print(matrix)
```
