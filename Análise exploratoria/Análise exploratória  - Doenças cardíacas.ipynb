```python
ax = sns.barplot(data = df, x="currentSmoker", y = "age");
ax.figure.set_size_inches(12, 8)
ax.set_title("Pessoas fulmantes", fontsize = 20);
ax.set_xlabel("Fulmantes", fontsize = 20);
ax.set_ylabel("Anos dos pacientes", fontsize = 20);
```

```python
ax = sns.boxplot(df["glucose"], orient='h', width=0.3);
ax.figure.set_size_inches(12, 8);
ax.set_title("Glicose dos pacientes", fontsize = 20);
ax.set_xlabel("Nivel de glicose", fontsize = 20);
```

```python
ax = sns.distplot(df["age"])
ax.figure.set_size_inches(12, 8)
ax.set_title("Idade dos pacientes", fontsize=20);
ax.set_ylabel('Idade', fontsize=20);
ax.set_xlabel("Idade por anos", fontsize=20)
ax
```

```python
ax = sns.countplot(df["age"]);
ax.figure.set_size_inches(12, 8);
ax.set_title("Idades dos pacientes", fontsize = 20);
ax.set_xlabel("Idades", fontsize = 20);
ax.set_ylabel("Contagem de idade", fontsize = 20);
```

```python
ax = sns.distplot(df["male"])
ax.figure.set_size_inches(12, 8)
ax.set_title("Pacientes masculino", fontsize=20);
ax.set_ylabel('Idade', fontsize=20);
ax.set_xlabel("Idade por anos", fontsize=20)
ax
```

```python
ax = sns.violinplot(data = df, x ="glucose");
ax.figure.set_size_inches(12, 8);
ax.set_title("Total de Glicose dos pacientes", fontsize = 20);
ax.set_xlabel("Contagem", fontsize = 20);
ax.set_ylabel("Glicose", fontsize = 20);
```

```python
plt.figure(figsize=(8, 6))

ax = sns.pairplot(df, hue = "currentSmoker",
                 vars = ["male", "age", "education", "diaBP", "BMI", "glucose"]);
```
